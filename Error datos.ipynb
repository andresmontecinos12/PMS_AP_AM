```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
libro="error_datos.xlsx"
datos=pd.read_excel(libro,sheet_name='OPERACIONES TIPO C')
x=datos[["Delta Año proyección","PCI Medido","centro pista","borde pista","plataforma","umbrales"]]
y=datos["Pci Proyectado1"]
df=pd.DataFrame(x)
```

```python
from sklearn.metrics import r2_score
print('Error predicción a 2 años:',r2_score(x['PCI Medido'].iloc[0:37],y[0:37]))
print('Error predicción a 3 años:',r2_score(x['PCI Medido'].iloc[38:77],y[38:77]))
```

```python
plt.scatter(x['PCI Medido'].iloc[0:37],y[0:37])
#plt.plot([y.min(), y.max()], [x['PCI Medido'].min(),x['PCI Medido'].max()], 'k--', lw=4)
plt.plot([0,100],[0,100],'k--', lw=4)
plt.xlabel('PCI Medido')
plt.ylabel('PCI Proyectado')
plt.text(20,80,'$R^2$=0.71')
plt.title('Predicciones a 2 años')
plt.show

```

```python
plt.scatter(x['PCI Medido'].iloc[38:77],y[38:77])
#plt.plot([y.min(), y.max()], [x['PCI Medido'].min(),x['PCI Medido'].max()], 'k--', lw=4)
plt.plot([0,100],[0,100],'k--', lw=4)
plt.xlabel('PCI Medido')
plt.ylabel('PCI Proyectado')
plt.text(20,80,'$R^2$=-0.49')
plt.title('Predicciones a 3 años')
plt.show

```

```python
fig, ax = plt.subplots()

ax.scatter(y[0:37], y[0:37]-x['PCI Medido'].iloc[0:37],y[0:37], edgecolors=(0, 0, 0),label='')
ax.plot([y[0:37].min(), y[0:37].max()], [0, 0], 'k--', lw=4,label='0')
ax.plot([y[0:37].min(), y[0:37].max()], [4, 4], 'r--', lw=2,label='+4')
ax.plot([y[0:37].min(), y[0:37].max()], [-4, -4], 'r--', lw=2,label='-4')
ax.set_xlabel('PCI Medido')
ax.set_ylabel('PCI Residuos del Modelo')
plt.ylim(-50, 50)
ax.set_title('Datos Proyecciones a 2 años')
plt.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(y[38:77], y[38:77]-x['PCI Medido'].iloc[38:77],y[38:77], edgecolors=(0, 0, 0),label='')
ax.plot([y[38:77].min(), y[38:77].max()], [0, 0], 'k--', lw=4,label='0')
ax.plot([y[38:77].min(), y[38:77].max()], [4, 4], 'r--', lw=2,label='+4')
ax.plot([y[38:77].min(), y[38:77].max()], [-4, -4], 'r--', lw=2,label='-4')
ax.set_xlabel('PCI Medido')
ax.set_ylabel('PCI Residuos del Modelo')
plt.ylim(-50, 50)
ax.set_title('Datos Proyecciones a 3 años')
plt.legend()

plt.show()
```
